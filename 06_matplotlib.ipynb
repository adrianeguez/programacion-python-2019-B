```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configuracion general de pyplot 
# (Revisar Doc)
plt.rcParams["figure.figsize"]= (4,3)
plt.rcParams["figure.dpi"] = 150
```

```python
valores_x = np.arange(0,10,0.1)
```

```python
valores_x
print()
```
